```python
import requests
import datetime

# Set your Eventbrite API token
EVENTBRITE_API_TOKEN = 'YJEYNTYK5QILPSAPJUG4'

# Define the API endpoint
EVENTBRITE_API_ENDPOINT = 'https://www.eventbriteapi.com/v3/events/search/'

# Get tomorrow's date and format it as a string in the Eventbrite API's date format
tomorrow = datetime.date.today() + datetime.timedelta(days=1)
tomorrow_str = tomorrow.strftime('%Y-%m-%dT00:00:00')

# Define the API parameters
EVENTBRITE_API_PARAMS = {
    'location.address': 'San Francisco',
    'location.within': '10mi',
    'start_date.range_start': tomorrow_str,
    'sort_by': 'date',
}

# Make the API call
response = requests.get(
    EVENTBRITE_API_ENDPOINT,
    params=EVENTBRITE_API_PARAMS,
    headers={
        'Authorization': f'Bearer {EVENTBRITE_API_TOKEN}',
        'Content-Type': 'application/json'
    },
)

# Parse the JSON response and extract the event information
events = response.json()['events']
for event in events:
    print(event['name']['text'])
    print(event['description']['text'])
    print(event['start']['local'])
    print('---')
```
