```python
import os
import pandas as pd

# Find all files starting with 'donyc_....csv' in current directory
files = [file for file in os.listdir('.') if file.startswith('donyc_') and file.endswith('.csv')]

# Retrieve the first file as donyc_df
donyc_df = pd.read_csv(files[0])
print(files)
for i in range(1, len(files)-1):
    #TODO Loop through remaining files and add 2nd column to donyc_df
    if i > 1:
        print(i)
        df = pd.read_csv(files[i])
        print(df.columns[1])
        donyc_df[df.columns[1]] = df.iloc[:, 1]

donyc_df.to_csv('total_donyc_events.csv', index=False)
```
