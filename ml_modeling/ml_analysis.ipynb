```python

# Encode the label using LabelEncoder
X = df.loc[:, 'Crime':].values
y = df.loc[:, 'Ridership'].values
# le = LabelEncoder()
# y = le.fit_transform(y)
 
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

#v Create a pipeline and fit the model
pipe_lr = make_pipeline(StandardScaler(),
                        PCA(n_components=2),
                        RandomForestClassifier(criterion='gini', n_estimators=50, max_depth=5, random_state=1))
 
pipe_lr.fit(X_train, y_train)
y_pred = pipe_lr.predict(X_test)
test_acc = pipe_lr.score(X_test, y_test)
print(f'Test accuracy: {test_acc:.3f}')
```

```python

# Encode the label using LabelEncoder
X = df.loc[:, 'Crime':].values
y = df.loc[:, 'Ridership'].values
# le = LabelEncoder()
# y = le.fit_transform(y)
 
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

pipe = make_pipeline(StandardScaler(), Ridge())
pipe.fit(X_train, y_train)
pipe.score(X_test, y_test)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.neighbors import KNeighborsRegressor

pipe = make_pipeline(StandardScaler(), KNeighborsRegressor())
pipe.fit(X_train, y_train)
pipe.score(X_test, y_test)
```

```python
from sklearn.model_selection import GridSearchCV

knn_pipe = make_pipeline(StandardScaler(), KNeighborsRegressor())
param_grid = {'kneighborsregressor__n_neighbors': range(1, 10)}
grid = GridSearchCV(knn_pipe, param_grid, cv=10)
grid.fit(X_train, y_train)
print(grid.best_params_)
print(grid.score(X_test, y_test))
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV

# Define the pipeline
knn_pipe = make_pipeline(StandardScaler(), KNeighborsRegressor())

# Define the hyperparameter grid to search over
param_grid = {'kneighborsregressor__n_neighbors': range(1, 10)}

# Perform a grid search over the hyperparameters using cross-validation
grid = GridSearchCV(knn_pipe, param_grid, cv=10)

# Fit the grid search to the training data
grid.fit(X_train, y_train)

# Print the best hyperparameters and test score
print("Best hyperparameters:", grid.best_params_)
print("Test score:", grid.score(X_test, y_test))
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV

# Define the pipeline
rf_pipe = make_pipeline(StandardScaler(), RandomForestRegressor(random_state=42))

# Define the hyperparameter grid to search over
param_grid = {
    'randomforestregressor__n_estimators': [50, 100, 150],
    'randomforestregressor__max_depth': [10, 20, 30, None],
}

# Perform a grid search over the hyperparameters using cross-validation
grid = GridSearchCV(rf_pipe, param_grid, cv=10)

# Fit the grid search to the training data
grid.fit(X_train, y_train)

# Print the best hyperparameters and test score
print("Best hyperparameters:", grid.best_params_)
print("Test score:", grid.score(X_test, y_test))
```
